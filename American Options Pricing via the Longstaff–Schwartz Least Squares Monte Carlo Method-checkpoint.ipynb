```python
import numpy as np
import math
import matplotlib.pyplot as plt
from dataclasses import dataclass
import json, time
from pathlib import Path

# where to dump the files (currently same folder)
save_dir = Path(".")

np.random.seed(0)   # reproducibility

@dataclass
class OptionSpec:
    S0: float
    K: float
    r: float
    q: float
    sigma: float
    T: float
    option_type: str = 'put'  # can be 'put' or 'call'

def crr_american_price(spec: OptionSpec, steps: int):
    """Simple CRR binomial tree for American options."""
    S0, K, r, q, sigma, T = spec.S0, spec.K, spec.r, spec.q, spec.sigma, spec.T
    dt = T / steps
    u = math.exp(sigma * math.sqrt(dt))
    d = 1 / u
    disc = math.exp(-r * dt)
    p = (math.exp((r - q) * dt) - d) / (u - d)

    # terminal stock prices
    stock_T = np.array([S0 * (u**j) * (d**(steps - j)) for j in range(steps + 1)])
    if spec.option_type.lower() == 'put':
        vals = np.maximum(K - stock_T, 0.0)
    else:
        vals = np.maximum(stock_T - K, 0.0)

    # work backwards
    for i in range(steps - 1, -1, -1):
        vals = disc * (p * vals[1:i+2] + (1 - p) * vals[0:i+1])
        stock_nodes = np.array([S0 * (u**j) * (d**(i - j)) for j in range(i + 1)])
        if spec.option_type.lower() == 'put':
            vals = np.maximum(vals, K - stock_nodes)
        else:
            vals = np.maximum(vals, stock_nodes - K)

    return vals[0]

def simulate_paths(S0, r, q, sigma, T, steps, n_paths):
    """Simulate GBM price paths using log returns."""
    dt = T / steps
    drift = (r - q - 0.5 * sigma**2) * dt
    vol_term = sigma * math.sqrt(dt)

    # normal shocks
    shocks = np.random.normal(loc=drift, scale=vol_term, size=(n_paths, steps))
    log_paths = np.log(S0) + np.cumsum(shocks, axis=1)
    # include the starting price
    log_paths = np.hstack([np.full((n_paths, 1), math.log(S0)), log_paths])

    return np.exp(log_paths)

def lsm_price(spec: OptionSpec, n_paths=20000, steps=50, degree=2, bermudan_exercise_times=None):
    """Longstaff–Schwartz method for American/ Bermudan options."""
    S0, K, r, q, sigma, T = spec.S0, spec.K, spec.r, spec.q, spec.sigma, spec.T
    dt = T / steps
    disc = math.exp(-r * dt)

    paths = simulate_paths(S0, r, q, sigma, T, steps, n_paths)

    if spec.option_type.lower() == 'put':
        payoff = np.maximum(K - paths, 0.0)
    else:
        payoff = np.maximum(paths - K, 0.0)

    cf = payoff[:, -1].copy()  # start from maturity
    exercise_flags = np.zeros_like(payoff, dtype=bool)
    exercise_flags[:, -1] = payoff[:, -1] > 0

    allowed_times = set(range(0, steps + 1)) if bermudan_exercise_times is None else set(bermudan_exercise_times)

    for t in range(steps - 1, 0, -1):
        if t not in allowed_times:
            cf *= disc
            continue

        itm = payoff[:, t] > 0
        if not np.any(itm):
            cf *= disc
            continue

        X = paths[itm, t]
        A = np.vstack([X**k for k in range(degree + 1)]).T
        Y = cf[itm] * disc
        coeffs, *_ = np.linalg.lstsq(A, Y, rcond=None)
        cont_val = A.dot(coeffs)

        ex_now = itm.copy()
        ex_now[itm] = payoff[itm, t] >= cont_val

        cf *= disc
        cf[ex_now] = payoff[ex_now, t]
        exercise_flags[ex_now, t] = True

    price = np.mean(cf * disc)
    stderr = np.std(cf * disc) / math.sqrt(n_paths)
    return price, stderr

def run_example():
    spec = OptionSpec(100.0, 100.0, 0.05, 0.0, 0.2, 1.0, option_type='put')

    step_sizes = [50, 100, 200, 400]
    crr_vals = []
    start_time = time.time()

    for N in step_sizes:
        crr_vals.append(crr_american_price(spec, N))

    lsm_val, lsm_err = lsm_price(spec)

    elapsed = time.time() - start_time

    print("CRR prices:", list(zip(step_sizes, crr_vals)))
    print(f"LSM estimate: {lsm_val:.6f} ± {lsm_err:.6f}")
    print(f"Run time: {elapsed:.3f} sec")

    # plot convergence
    plt.figure(figsize=(7, 4))
    plt.plot(step_sizes, crr_vals, marker='o', label='CRR American')
    plt.hlines(lsm_val, step_sizes[0], step_sizes[-1], colors='k', linestyles='--', label='LSM')
    plt.xlabel("Steps")
    plt.ylabel("Price")
    plt.title("CRR Convergence vs LSM")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # save report
    output_data = {
        "spec": spec.__dict__,
        "crr": list(zip(step_sizes, crr_vals)),
        "lsm": {"price": lsm_val, "stderr": lsm_err}
    }
    with open(save_dir / "report.json", "w") as f:
        json.dump(output_data, f, indent=2)

if __name__ == "__main__":
    run_example()
```
